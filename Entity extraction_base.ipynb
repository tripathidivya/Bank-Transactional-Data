```python
import pandas as pd
import numpy as np
```

```python
statement_df= pd.read_csv("EOD_Transactions.csv")
```

```python
statement_df.head()
```

```python
import re

def extract_entities(narration):
    if isinstance(narration, str):
        entities = re.findall(r'\b[A-Z]+\b', narration)
        return entities
    else:
        return []
```

```python
# Apply the function to the 'Narration' column
statement_df['Entities'] = statement_df['Narration'].apply(extract_entities)
```
